```python
n='Rodolfo Jose Maria Salas Lopez'
def getlongnames(n,nn={}):
    import re
    ilast=100 #just a large integer
    pos=[]
    i=0
    if len(n.split())<=4:
        print('WARNING: Expected long name with at most of 4 words')
    for p in n.split():
        i=i+1
        if re.search('\w',p, re.UNICODE):
            if len(p)<=3:
                ilast=i
                pos.append(i)
    if len(pos)>1 and (pd.np.array(pos[1:])-pd.np.array(pos[:-1])).max()>1:
        #special cases
        ilast=pos[0]
    if ilast<3:
        nn['NOMBRES']  =(' ').join( n.split()[:3] )
        nn['APELLIDOS']=(' ').join( n.split()[3:])
    elif ilast<4 and len(pos)>1:
        nn['NOMBRES']  =(' ').join( n.split()[:4] )
        nn['APELLIDOS']=(' ').join( n.split()[4:])
    else:
        nn['NOMBRES']  =(' ').join( n.split()[:2] )
        nn['APELLIDOS']=(' ').join( n.split()[2:])
    
    return nn    
```

```python
def getnames(n):
    nn={}
    if len(n.split())==4:
        nn['NOMBRES']  =(' ').join( n.split()[:2] )
        nn['APELLIDOS']=(' ').join( n.split()[-2:] )
    elif len(n.split())==3:
        nn['NOMBRES']  =(' ').join( n.split()[:2] )
        nn['APELLIDOS']=(' ').join( n.split()[-1:] )
    elif len(n.split())==2:
        nn['NOMBRES']  =(' ').join( n.split()[:1] )
        nn['APELLIDOS']=(' ').join( n.split()[-1:] )
    elif len(n.split())>4:
        nn=getlongnames(n,nn)
    else:
        nn=None
    return nn
```

```python
li['NOMBRES Y APELLIDOS']=li['NOMBRE COMPLETO'].apply( getnames )
#li['APELLIDOS']=li['NOMBRE COMPLETO'].apply( getlastnames )
```
